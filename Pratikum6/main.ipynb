```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import scipy.cluster.hierarchy as sch
```

```python
# Membaca data
url = "D:\Coding\Pembelajaran Mesin\Pembelajaran_Mesin\Pratikum6\Data\heart_failure_clinical_records_dataset.csv"
data = pd.read_csv(url)
print(data)
```

```python
# Import libarary PCA model
from sklearn.decomposition import PCA

# Inisialisasi Class
pca = PCA(n_components=2)
```

```python
#  Fliting the feature into PCA model
dataBaru = pca.fit_transform(data)
print(dataBaru)
```

```python
# Eigen vector (v1 & v2)
pca.explained_variance_ratio_
```

```python
dendogramData = sch.linkage(data, method='single', metric='cityblock')
sch.dendrogram(Z=dendogramData)
plt.show()
```

```python
dendogramDataBaru = sch.linkage(dataBaru, method='single', metric='cityblock')
sch.dendrogram(Z=dendogramDataBaru)
plt.show()
```
